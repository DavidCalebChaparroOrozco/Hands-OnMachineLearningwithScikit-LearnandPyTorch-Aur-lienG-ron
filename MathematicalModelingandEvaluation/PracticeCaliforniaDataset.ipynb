$$
\mathrm{RMSE}(X, y, h) \;=\; \sqrt{\frac{1}{m} \sum_{i=1}^{m} \left( h\left(x^{(i)}\right) - y^{(i)} \right)^2}
$$


```python
def rmse(X, y, h):
    m = len(y)
    error_sum = 0

    for i in range(m): 
        y_pred = h(X[i]) # 4.528 - 4.52 = 0.008
        error_sum += (y_pred - y[i]) ** 2 # (0.008) ** 2 = 0.000064

    return np.sqrt(error_sum / m) # sqrt(0.000064 / m)
```

$$
\mathrm{MAE}(X, y, h) \;=\; \frac{1}{m} \sum_{i=1}^{m} \left| h\left(x^{(i)}\right) - y^{(i)} \right|
$$

```python
def mae (X, y, h):
    m = len(y)
    error_sum = 0

    for i in range(m):
        y_pred = h(X[i]) # 4.528
        error_sum += abs(y_pred - y[i]) # abs(4.528 - 4.52) = 0.008

        return error_sum / m # 0.008 / m
```
